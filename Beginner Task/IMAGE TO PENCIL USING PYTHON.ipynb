# Importing libraries

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading the image


```python
img=cv2.imread("rocky.jpg")
img=cv2.resize(img,(600,900))
i_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(i_rgb)
```

# Converting Image to grayscale

```python
from IPython.display import display
```

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grey_img))
```

# Converting  image to Negative

```python
invt_img = cv2.bitwise_not(grey_img)
display(Image.fromarray(invt_img))
```

# Smoothening the image

```python
blur_img=cv2.GaussianBlur(invt_img,(5,5),0,0)
display(Image.fromarray(blur_img))
```

# Creating the pencil Sketch

```python
invt_blur_img=255-blur_img
sketch=cv2.divide(grey_img,invt_blur_img,scale=250.0)
display(Image.fromarray(sketch))
```

# Comparing the sketch and the Image

```python
i_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(i_rgb)
display(Image.fromarray(sketch))
```
